```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
img = cv2.imread("Images/pattern_img.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize = (25,25))
plt.imshow(img, cmap='gray')
```

```python
from ip_functions import threshold_img

def bit_plane_slicing(img):
    bit_planes = [1,2,4,8,16,32,64,128]
    
    fig, axes = plt.subplots(nrows = 4, ncols = 2, figsize = (24,24))
    axes[0,0].imshow(img, cmap='gray')
    row_count = 0
    col_count = 0
    for i, bit in enumerate(bit_planes):
        axes[row_count, col_count].imshow(np.bitwise_and(bit, img), cmap='gray')
        axes[row_count, col_count].set_title(f"Bit={bit}")
        col_count += 1
        if col_count%2 == 0:
            col_count = 0
            row_count += 1    
```

```python
bit_plane_slicing(img)
```
